# CLASS EDE

Early dark energy is realized through a scalar field with the following potential,  
  
V($\phi$) = $(mf)^2\left(1 - \cos\left(\frac{\phi}{f}\right)\right)^n + \Lambda$,  
  
where $m$ is the axion mass, $f$ the axion decay constant, $\phi$ the scalar field, and $\Lambda$ is the cosmological constant.

```python
# Import relevant modules
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from classy import Class
```

One can, as in standard CLASS, set values for various parameters as follows.

```python
M = Class()
M.set({'A_s': 2.163e-09,
       'n_s': 0.9843,
       '100*theta_s': 1.0415361456812424,
       'omega_b': 0.022383483553045907,
       'omega_cdm': 0.1247,
       'tau_reio': 0.0656,
       'N_ncdm':1,
       'N_ur':2.0328})
```

The next block of code is an example of how to implement EDE. Firstly, recall that the cosmological constant is included in the potential of the scalar field. We do this so that it is the quatity which is tuned when trying to match the budget equation. The various parameters for the scalar field potential have been added as seperate variables (abandoning scf_parameters for easy interface with MCMC.)

```python
M.set({'Omega_Lambda':0,
       'Omega_fld':0,
       'Omega_scf':-1,
       'scf_parameters':'1, 1, 1, 1, 1, 0.0',
       'scf_tuning_index':3,
       'attractor_ic_scf':'no',
       'output':'tCl pCl lCl',
       'n_scf':3, 
       'CC_scf':1, 
       'f_scf': 4.383e+26, 
       'm_scf': 3.4e-27,
       'thetai_scf': 2.64})
```

```python
M.compute()
baM = M.get_background()
```

```python
fEDE = M.fEDE()
z_c = M.z_c()
baH = baM['H [1/Mpc]']
baT = baM['conf. time [Mpc]']
baa = 1/(1 + baM['z'])
bV = baM['V_e_scf']
bpp = baM["phi'_scf"]
baCrit = baM['(.)rho_crit']
rho_scf = (bpp*bpp/(2*baa*baa) + bV)/3.
plt.figure(figsize=(10,8))
plt.xscale('log')
plt.xlim([10,1e5])
plt.xlabel('z')
plt.ylabel(r'$f_{EDE}$')
plt.rcParams.update({'font.size': 18})
plt.plot(baM['z'],rho_scf/baCrit,lw=2,c='k')
```

```python
def_params = {'A_s': 2.163e-09,
       'n_s': 0.9843,
       '100*theta_s': 1.0415361456812424,
       'omega_b': 0.022383483553045907,
       'omega_cdm': 0.1247,
       'tau_reio': 0.0656,
       'N_ncdm':1,
       'N_ur':2.0328,
       'Omega_Lambda':0,
       'Omega_fld':0,
       'Omega_scf':-1,
       'scf_parameters':'1, 1, 1, 1, 1, 0.0',
       'scf_tuning_index':3,
       'attractor_ic_scf':'no',
       'output':'tCl pCl lCl',
       'CC_scf':1, 
       'f_scf': 4.383e+26, 
       'm_scf': 3.4e-27,
       'thetai_scf': 2.64}
```

```python
M3, M4, M5 = Class(), Class(), Class()

M3.set(def_params)
M3.set({'n_scf':3})
M3.compute()
baM3 = M3.get_background()

M4.set(def_params)
M4.set({'n_scf':4})
M4.compute()
baM4 = M4.get_background()

M5.set(def_params)
M5.set({'n_scf':5})
M5.compute()
baM5 = M5.get_background()
```

```python
baH3 = baM3['H [1/Mpc]']
baT3 = baM3['conf. time [Mpc]']
baa3 = 1/(1 + baM3['z'])
bV3 = baM3['V_e_scf']
bpp3 = baM3["phi'_scf"]
baCrit3 = baM3['(.)rho_crit']
rho_scf3 = (bpp3*bpp3/(2*baa3*baa3) + bV3)/3.


baH4 = baM4['H [1/Mpc]']
baT4 = baM4['conf. time [Mpc]']
baa4 = 1/(1 + baM4['z'])
bV4 = baM4['V_e_scf']
bpp4 = baM4["phi'_scf"]
baCrit4 = baM4['(.)rho_crit']
rho_scf4 = (bpp4*bpp4/(2*baa4*baa4) + bV4)/3.


baH5 = baM5['H [1/Mpc]']
baT5 = baM5['conf. time [Mpc]']
baa5 = 1/(1 + baM5['z'])
bV5 = baM5['V_e_scf']
bpp5 = baM5["phi'_scf"]
baCrit5 = baM5['(.)rho_crit']
rho_scf5 = (bpp5*bpp5/(2*baa5*baa5) + bV5)/3.


plt.figure(figsize=(10,8))
plt.xscale('log')
plt.xlim([10,1e5])
plt.xlabel('z')
plt.ylabel(r'$f_{EDE}$')
plt.rcParams.update({'font.size': 18})
plt.plot(baM3['z'],rho_scf3/baCrit3,lw=2,c='r',label='n = 3')
plt.plot(baM4['z'],rho_scf4/baCrit4,lw=2,c='b',label='n = 4')
plt.plot(baM5['z'],rho_scf5/baCrit5,lw=2,c='g',label=r'n = 5')
plt.legend(loc='upper left')
```
